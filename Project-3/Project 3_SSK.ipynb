### Model 1: Logistic Regression

**Logistic regression** is a useful algorithm for classification. Here, the outcome has been turned into a binary variable taking the value of 0 (Pass) or 1 (Fail). Having a binary outcome one of the assumptions of logistic regression, so that makes it a great starting point and a plausible choice for use in this case! The other assumptions of logistic regression are similar to other regression models (e.g., independent observations, normality, etc.), so that is great for building off of the previous project.

Logistic regression is a pretty simple model and a lot of people have familiarity with it, so that is a big pro. One of the cons is that it can be difficult to interpret. Another model may be better for solving this particular problem but I like the idea of starting with logistic regression as a parsimonious baseline.

#### Train Model

```python
# import library from sklearn
from sklearn.linear_model import LogisticRegression

# import libraries for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore')
from sklearn.metrics import accuracy_score
```

```python
# create a model
logit_reg = LogisticRegression()

# fit the model
logit_model = logit_reg.fit(X, y)

# predict the outcome 
y_pred = logit_model.predict(X)
```

#### Hyperparameter Tuning

```python
# set parameters
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
             'C': np.arange(.1, 1, .1),
             'fit_intercept': [True, False],
             'solver': ['liblinear', 'saga']}


# execute the grid search and fit to data
logit_grid = GridSearchCV(logit_model, 
                          param_grid, 
                          cv=3)
logit_grid.fit(X, 
               y)

# choose best performing model
best_index = np.argmax(logit_grid.cv_results_["mean_test_score"])
best_logit_pred = logit_grid.best_estimator_.predict(X)

# print results
print(logit_grid.cv_results_["params"][best_index])
print('Validation Accuracy', accuracy_score(best_logit_pred, y))
```

#### Re-fit Model

```python
# re-create a model
logit_reg2 = LogisticRegression(C=0.1)

# re-fit model 
logit_model2 = logit_reg2.fit(X, y)

# predict the outcome 
y_pred = logit_model2.predict(X)
```

#### Confusion Matrix

```python
# import libraries
from sklearn.metrics import confusion_matrix

# create matrix
cf_matrix = confusion_matrix(y, 
                             y_pred, 
                             normalize = "true")

# convert to a dataframe
df_cm = pd.DataFrame(cf_matrix, range(2), range(2))

# set figure specifications
df_cm = df_cm.rename(index=str, columns={0: "Fail", 1: "Pass"})
df_cm.index = ["Fail", "Pass"]
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)#for label size
sns.heatmap(df_cm, 
           annot=True,
           annot_kws={"size": 16},
           fmt='g')

# plot
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

#### Validation Metrics

```python
# set each object to 0 to start
TP = 0
FP = 0
TN = 0
FN = 0

# loop over each observation to assign to TP, FP, TN, and FN
for i in range(len(y_pred)): 
    if y[i]==y_pred[i]==1:
        TP += 1
    if y_pred[i]==1 and y[i]==0:
        FP += 1
    if y[i]==y_pred[i]==0:
        TN += 1
    if y_pred[i]==0 and y_pred[i]!=y[i]:
        FN += 1
```

```python
# accuracy
accuracy = (TP + TN)/(TP + TN + FP + FN)
print("Accuracy for Model 1 (Logistic Regression) is", accuracy)
```

```python
# recall
recall = TP/(TP + FN)
print("Recall for Model 1 (Logistic Regression) is", recall)
```

```python
# precision
precision = TP/(TP + FP)
print("Precision for Model 1 (Logistic Regression) is", precision)
```

```python
# f1 score 
f1 = 2 * (precision * recall)/(precision + recall)
print("F1 Score for Model 1 (Logistic Regression) is", f1)
```

###  Model 2: Random Forests

**Random Forests** is a decision tree model that involves many trees, using a subset of features to grow each tree. The trees then make predictions by using the majority of the trees in making decisions. In Random Forests, each tree is trained independently and then are combined. Random Forests tend to be a pretty accurate categorication model which is a big *pro*. Our outcome here is only binary so we may not see a big benefit using Random Forests, but it's worth an investigation! A *con* of this approach is that Random Forests are prone to overfitting. As opposed to logistic regression (Model 1), Random Forests doesn't have any assumptions about the distribution of data which is nice!

```python
# import libraries from sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate, cross_val_score
from sklearn import tree
from sklearn.metrics import make_scorer, accuracy_score, recall_score, precision_score, f1_score
```

#### Fit Model

```python
# initialize a random forest classifier
rf_classifier = RandomForestClassifier(
                       # specify parameters
                       n_estimators=100,              # specify the number of trees (100 is default)
                       criterion='gini',              # or you can use 'entropy' for information gain
                       max_depth=None,                # how deep tree nodes can go
                       min_samples_split=2,           # samples needed to split node
                       min_samples_leaf=1,            # samples needed for a leaf
                       min_weight_fraction_leaf=0.0,  # weight of samples needed for a node
                       max_features=None,             # number of features to look for when splitting
                       max_leaf_nodes=None,           # max nodes
                       min_impurity_decrease=1e-07,   # early stopping
                       random_state = 10)             # random seed
```

#### Cross-Validation

```python
# specify cross-validation
scores = cross_val_score(rf_classifier, 
                         X, 
                         y.ravel(), 
                         cv=5)
```

```python
# calculate the average score across models
scores.mean()
```

#### Feature Importance

```python
# fit the random forest on data to get feature importance
# ----------
rf_classifier.fit(X, y.ravel())

# create feature importance dataframe
feat_importances = pd.concat([pd.DataFrame(X.columns),pd.DataFrame(np.transpose(rf_classifier.feature_importances_))], axis = 1)
feat_importances.columns = ["Feature", "Importance"]

# plot 
sns.barplot(x = "Importance", 
            y = "Feature", 
            data = feat_importances.nlargest(20, 'Importance')) # identify the 10 most important features
plt.show()
```

#### Re-fit Model

```python
# refit a basic decision tree using reduced number of features 
# ----------
dt_reduced_classifier = tree.DecisionTreeClassifier(
                                 # specify parameters
                                 criterion='gini',              # or 'entropy' for information gain
                                 splitter='best',               # or 'random' for random best split
                                 max_depth=5,                   # how deep tree nodes can go
                                 min_samples_split=2,           # samples needed to split node
                                 min_samples_leaf=1,            # samples needed for a leaf
                                 min_weight_fraction_leaf=0.0,  # weight of samples needed for a node
                                 max_features=None,             # number of features to look for when splitting
                                 max_leaf_nodes=None,           # max nodes
                                 min_impurity_decrease=1e-07,   # early stopping
                                 random_state = 10)             # random seed

# pull out the 14 most features
important_features = feat_importances.nlargest(14, 'Importance')['Feature']

# create new dataset with only most important features
X_reduced = X[X.columns[X.columns.isin(important_features)]]

# fit the model on the new reduced model
dt_reduced_classifier.fit(X_reduced,
                          y)

# set column names as list
reduced_column_names = X_reduced.columns.tolist()

# make predictions 
forests_pred = dt_reduced_classifier.predict(X_reduced)
```

#### Confusion Matrix

```python
# create matrix
cf_matrix = confusion_matrix(y, 
                             forests_pred, 
                             normalize = "true")

# convert to a dataframe
df_cm = pd.DataFrame(cf_matrix, range(2), range(2))

# set figure specifications
df_cm = df_cm.rename(index=str, columns={0: "Fail", 1: "Pass"})
df_cm.index = ["Fail", "Pass"]
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)#for label size
sns.heatmap(df_cm, 
           annot=True,
           annot_kws={"size": 16},
           fmt='g')

# plot
plt.title("Confusion Matrix: Model 2")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

#### Validation Metrics

```python
# accuracy

accuracy = accuracy_score(y, forests_pred)
print("Accuracy for Model 2 (Random Forests) is", accuracy)
```

```python
# recall

recall = recall_score(y, forests_pred, average='weighted')
print("Recall for Model 2 (Random Forests) is", recall)
```

```python
# precision

precision = precision_score(y, forests_pred, average='weighted')
print("Precision for Model 2 (Random Forests) is", precision)
```

```python
# f1 score 

f1 = f1_score(y, forests_pred, average='weighted')
print("F1 Score for Model 2 (Random Forests) is", f1)
```

### Model 3: XG Boost 

**XG Boost** is another tree-based approach. Building off of Random Forests (where trees are independently trained), XG Boost (and other boosting methods) start with a strong classifier and iterate on it with weaker ones. Those weak learners are trained on errors. Because of this process, XG Boost is great for challenging classification problems, which is definitely a *pro*. However, this process also naturally lends itself to the *con* of potentially overfitting. Another *pro* compared to other boosting methods is that XG Boosts are relatively computationally inexpensive. 

```python
# import libraries
!pip install xgboost
import xgboost as xgb 
```

#### Train Model + Cross-Validation

```python
# initialize an XGBoost classifier
# ----------
xgb_classifier = xgb.XGBClassifier(random_state=10)

# define the scoring metrics
scoring = {
          'accuracy': make_scorer(accuracy_score),
          'recall': make_scorer(recall_score),
          'precision': make_scorer(precision_score),
          'f1': make_scorer(f1_score)
          }


# perform cross-validation with 5-fold and return the trained estimators
cv_results = cross_validate(xgb_classifier,        # specify estimator 
                            X,                     # specify features
                            y.ravel(),             # specify outcome, and use ravel
                            cv=5,                  # specify 5-fold cross validation
                            return_estimator=True, # return the estimators fitted at each split
                            scoring=scoring)       # which scoring metrics to return (the whole list in this case)
```

#### Validation Metrics

```python
# print the results for accuracy, recall, precision, and F1 score
print("Validation Metrics for Model 3 (XG Boost) are:") 
for metric in ['test_accuracy', 'test_recall', 'test_precision', 'test_f1']:
    scores = cv_results[metric]
    print(f"{metric[5:]} - {scores.mean():.3f}")
```

*Q1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?*

**ANSWER:**

Accuracy gives a quick glance at how well a model is doing but I think that other metrics are also important to consider depending on the broader context surrounding a particular research question / model. In some cases, it may be preferred to have more false positives because the stakes are high - ascertaining health risk and making decisions about biopsies or recommending additional medical testing may be one of these situations. In other cases, it may be preferred to get a positive categorization *only* when the categorization is actually positive (in other words, to reduce false positives). 

Taken together, thinking critically about the research question and the other circumstances at play is importat when employing machine learning in social sciences and public policy. Machine learning feels like magic, but it isn't! It requires careful consideration from the humans who are both building and evaluating the models.
