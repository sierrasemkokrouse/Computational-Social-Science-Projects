### Train PCA and Discuss Number of Components

*I'm first going to fit a PCA on the entire dataset and create a scree plot to determine how many components to choose. The scree plot will show both individual variance and cumulative variance. Individual variance is the variance explained by each component. Cumulative variance is the the running total of variance as each component is added.* 

```python
# Fit PCA on full data 
# ----------
pca_all = PCA() 
principalComponents_nhanes = pca_all.fit_transform(nhanes_scaled)

# Create scree plot 
# ----------
PC_values = np.arange(pca_all.n_components_) + 1
cumulative_variance_ratio = np.cumsum(pca_all.explained_variance_ratio_)

plt.plot(PC_values, pca_all.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.plot(PC_values, cumulative_variance_ratio, 'ro-', linewidth=2, color='red')
plt.title('Scree Plot with All Components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.legend(['Individual Variance Ratio', 'Cumulative Variance Ratio'])

ax = plt.gca()
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_ylabel('Variance Explained (%)')
ax2.set_yticklabels(['{:,.0%}'.format(x) for x in ax2.get_yticks()])

plt.show()
```

*It is helpful to see all features here but there are so many features (about 250 total) that the individual variance curve is difficult to see. It looks like the slope of cumulative variance starts to taper off at 100 components, so I'm going to look at 100 components next.*

```python
# Train PCA with 100 components
# ----------
pca_100 = PCA(n_components=100)
principalComponents_100 = pca_100.fit_transform(nhanes_scaled)

# Create scree plot
# ----------
PC_values_100 = np.arange(pca_100.n_components_) + 1
cumulative_variance_ratio = np.cumsum(pca_100.explained_variance_ratio_)

plt.plot(PC_values_100, pca_100.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.plot(PC_values_100, cumulative_variance_ratio, 'ro-', linewidth=2, color='red')
plt.title('Scree Plot with 100 Components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.legend(['Individual Variance Ratio', 'Cumulative Variance Ratio'])

ax = plt.gca()
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_ylabel('Variance Explained (%)')
ax2.set_yticklabels(['{:,.0%}'.format(x) for x in ax2.get_yticks()])

plt.show()
```

*I'm next going to try to look at just 50 components. I feel that this might be a good middle ground between sufficient variance explained and parismony!*

```python
# Train PCA with 50 components
# ----------
pca_50 = PCA(n_components=50)
principalComponents_50 = pca_50.fit_transform(nhanes_scaled)

# Create scree plot 
# ----------
PC_values_50 = np.arange(pca_50.n_components_) + 1
cumulative_variance_ratio = np.cumsum(pca_50.explained_variance_ratio_)

plt.plot(PC_values_50, pca_50.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.plot(PC_values_50, cumulative_variance_ratio, 'ro-', linewidth=2, color='red')
plt.title('Scree Plot with 50 Components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.legend(['Individual Variance Ratio', 'Cumulative Variance Ratio'])

ax = plt.gca()
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_ylabel('Variance Explained (%)')
ax2.set_yticklabels(['{:,.0%}'.format(x) for x in ax2.get_yticks()])

plt.show()
```

#### **Final Choice of Number of Components:**

### Choose a Clustering Algorithm

#### **Answer:** 

*I'm going to use the k-means clustering algorithm. This is an unsupervised learning algorithm will group the data based on the distance between individual points and the center of the clusters. In this process, the algorithm attempts to minimize within cluster variation by minimizing the within-cluster sum of squares (inertia). You do have to specify the number of clusters beforehand, but we can again use a scree plot to determine how many clusters may be optimal.*

```python
# set random seed
np.random.seed(28)

k_values = range(1, 21)

# calculate inertia for different values of k
inertia = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(nhanes_scaled)
    inertia.append(kmeans.inertia_)

# plot the elbow method using inertia_
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_values)
plt.grid()
plt.show()
```

```python
# Set random seed
# ----------
np.random.seed(28)

# Use the inertia_ attribute to create the elbow plot
# ----------
inertia = []
k_values = range(1, 21)
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(nhanes_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow method to choose k
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k using inertia_')
plt.xticks(k_values)
plt.grid()
plt.show()
```

#### **Explanation:** 

*The plot above shows the number of clusters (1 to 20) on the x-axis compared to inertia (or the sum of squared distances from each data point to the cluster center) on the y-axis. This plot can be used to perform the elbow method where the rate of decrease in inertia changes most dramatically and where additional clusters would not provide much additional benefit. Using the elbow method, I'm going to proceed with 4 clusters. There is a large decrease in inertia from 3 to 4 clusters, and the intertia levels off from there to 20 clusters. Thus, including more than 4 clusters would not add any additional benefit.* 

### Cluster nhanes

The goal of PCA is dimensionality reduction, or reducing the number of features in a model. In clustering, on the other hand, the goal is to cluster data points together based on dis/similarity. The new features here are actually groups of data points.*
